# Reproducing plots in GMT Docs with Python.  

In this notebook (and the GMT_Chapter_13_examples one too), we work through the GMT Document examples, re-creating each plot with only python.  

```python
from mpl_toolkits.basemap import Basemap
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
# this lets you plot in the notebook
%matplotlib inline 


```

### Albers equal-area conic section. (GMT Docs Example 12.1.1)

```python
# setup albers equal area conic basemap
# lat_1 is first standard parallel.
# lat_2 is second standard parallel.
# lon_0,lat_0 is central point.
m = Basemap(resolution='l',projection='aea',\
            llcrnrlon=110,llcrnrlat=19,urcrnrlon=140,urcrnrlat=35,lon_0=125,lat_0=25)
m.drawcoastlines()
m.fillcontinents(color='lightgreen')
# labels = [left,right,top,bottom]
m.drawparallels(np.arange(20.,40,10.),labels=[True,True,False,False])
m.drawmeridians(np.arange(110,150.,10.),labels=[False,False,True,True])
# Basemap doesn't do the curved boundaries, but otherwise this looks
# good. 
```

### Equidistant conic (GMT Docs Example 12.1.2)

```python
# setup equidistant conic basemap.
# lat_1 is first standard parallel.
# lat_2 is second standard parallel.
# lon_0,lat_0 is central point.
# resolution = 'l' for low-resolution coastlines.
m = Basemap(width=12000000,height=9000000,
            resolution='l',projection='eqdc',\
            llcrnrlon=-88,llcrnrlat=18,urcrnrlon=-70,urcrnrlat=24,lat_0=21,lon_0=-79.)
m.drawcoastlines() # put on the coastlines
m.drawcountries() # and the countries
m.fillcontinents(color='lightgreen') # make land lightgreen
# draw parallels and meridians.
m.drawparallels(np.arange(20.,25.,5.),labels=[True,True,False,False])
m.drawmeridians(np.arange(-85.,-65,5.),labels=[False,False,True,True])

```

### Lambert conformal conic (GMT Docs Example 12.1.3)

```python
# set up the Lambert conformal conic projection (lcc)
m = Basemap(resolution='l',projection='lcc',\
            llcrnrlon=-130,llcrnrlat=24,urcrnrlon=-70,urcrnrlat=52,\
           lat_0=33,lon_0=-100)
m.drawcoastlines()
m.drawcountries(color='red',linewidth=2) # make the country boundarys fat red
m.drawstates() # put on the states
m.drawmapboundary(fill_color='blue') # color in the ocean blue

m.fillcontinents(color='tan',lake_color='blue') # make the land tan
# draw parallels and meridians.
m.drawparallels(np.arange(30.,50.,15.),labels=[True,True,False,False])
m.drawmeridians(np.arange(-120.,-65,15.),labels=[False,False,True,True])

```

### Lambert Azimuthal Equal Area (GMT Docs Example 12.2.1)

```python
# setup lambert azimuthal equal area basemap.
# lat_ts is latitude of true scale.
# lon_0,lat_0 is central point.
m = Basemap(llcrnrlon=0,llcrnrlat=-40,urcrnrlon=60.,urcrnrlat=-10,\
            resolution='l',projection='laea',\
            lat_ts=-30,lat_0=-30,lon_0=30)
m.drawcoastlines(linewidth=.5)
m.fillcontinents(color='lightgrey') # make the land light grey
# draw parallels and meridians.
m.drawparallels(np.arange(-30,-10,15.),labels=[True,True,False,False])
m.drawmeridians(np.arange(0,70.,15.),labels=[False,False,True,True])
m.drawmapboundary(fill_color='white') # make the ocean white
```

###  Polar stereographic projection (GMT Docs Example 12.2.2)
- doesn't work very well...

```python
# setup north polar stereographic basemap.
# The longitude lon_0 is at 6-o'clock, and the
# latitude circle boundinglat is tangent to the edge
# of the map at lon_0. Default value of lat_ts
# (latitude of true scale) is pole.
m = Basemap(llcrnrlon=-30,llcrnrlat=60,urcrnrlon=30,urcrnrlat=75,\
            projection='npstere',boundinglat=60,lon_0=0,resolution='l',round=True)
m.drawcoastlines()
m.fillcontinents(color='blue')
m.drawmapboundary(fill_color='ivory')
# draw parallels and meridians.
m.drawparallels(np.arange(60,75,10.),labels=[True,True,False,False])
m.drawmeridians(np.arange(-30,40.,10.),labels=[False,False,True,True])

```

### Simple Mercator  (GMT Docs Example 12.3.1)

```python
# llcrnrlat,llcrnrlon,urcrnrlat,urcrnrlon
# are the lat/lon values of the lower left and upper right corners
# of the map.
# lat_ts is the latitude of true scale.
# resolution = 'c' means use crude resolution coastlines.
m = Basemap(projection='merc',llcrnrlat=-70,urcrnrlat=70,\
            llcrnrlon=0,urcrnrlon=360,lat_ts=0,resolution='l')
#m.drawcoastlines()
m.fillcontinents(color='red')
# draw parallels and meridians.
m.drawparallels(np.arange(-90.,91.,30.),labels=[True,True,False,False])
m.drawmeridians(np.arange(-180.,181.,60.),labels=[False,False,False,True])
m.drawmapboundary(fill_color='white')
plt.title("Mercator Projection")
```

### Rectangular Transverse Mercator  (GMT Docs Example 12.3.2)

```python
# llcrnrlat,llcrnrlon,urcrnrlat,urcrnrlon
# are the lat/lon values of the lower left and upper right corners
# of the map.
# resolution = 'i' means use intermediate resolution coastlines.
# lon_0, lat_0 are the central longitude and latitude of the projection.
m = Basemap(llcrnrlon=20,llcrnrlat=29,urcrnrlon=50,urcrnrlat=45,\
            resolution='i',projection='tmerc',lon_0=35,lat_0=35)
m.drawcoastlines(linewidth=.5)
m.fillcontinents(color='orange',lake_color='seashell')
# draw parallels and meridians.
m.drawparallels(np.arange(30,50.,10),labels=[True,True,False,False])
m.drawmeridians(np.arange(20.,50.,10),labels=[False,False,False,True])
m.drawmapboundary(fill_color='seashell')
plt.title("Transverse Mercator Projection")
```

### Oblique Mercator  (GMT Docs Example 12.3.4)

Oblique mercator doesn't do the 'oblique' part.  Bug in basemap? 

```python
# setup oblique mercator basemap.
# width is width of map projection region in km (xmax-xmin_
# height is height of map projection region in km (ymax-ymin)
# lon_0, lat_0 are the central longitude and latitude of the projection.
# lat_1,lon_1 and lat_2,lon_2  are two pairs of points that define
# the projection centerline.
# Map projection coordinates are automatically rotated to true north.
# To avoid this, set no_rot=True.
# area_thresh=1000 means don't plot coastline features less
# than 1000 km^2 in area.
m = Basemap(height=2000000,width=3500000,\
            resolution='l',area_thresh=1000.,projection='omerc',\
            lon_2=285,lat_2=15,lon_1=295,lat_1=30,lon_0=-75,lat_0=25)
m.fillcontinents(color='tan',lake_color='lightblue')
m.drawcoastlines(linewidth=.5)
# draw parallels and meridians.
m.drawparallels(np.arange(-80.,81.,10.))
m.drawmeridians(np.arange(-180.,181.,10.))
m.drawmapboundary(fill_color='lightblue')
plt.title("Oblique Mercator Projection")
```
